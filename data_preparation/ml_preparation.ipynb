## Change player birthday_date to age
We use match date and player birthday date to calculate player age, which is important factor in player performance

```python
birthday_cols = df.columns.values[56::44]
birthday_cols
```

```python
len(birthday_cols)
```

```python
index = 56
for birthday_col in birthday_cols:
    # Calculate new value
    new_col_name = birthday_col.replace('birthday', 'age')
    df[new_col_name] = (df['date'] - df[birthday_col]).astype('timedelta64[Y]').astype('int64')
    index += 44
```

Now we can remove columns which we do not need as: players birthday columns

```python
df.drop(birthday_cols, axis=1, inplace=True)
```

```python
df.head()
```

There is only one object left and that is the date. To use it for learning, we can transform it into 2 columns: month and year of the match. We do not need day of match, because in our opinion it has no effect on match. But month can has impact on result, because the air temperature depends on the month. We also leave the year of the match being played, because football keeps changing, maybe our neural network will learn how football evaulates and will adjust to it.



But for now we drop it

```python
df['month'] = df['date'].astype('str').str[5:7].astype('int64')
df['year'] = df['date'].astype('str').str[:4].astype('int64')
```
